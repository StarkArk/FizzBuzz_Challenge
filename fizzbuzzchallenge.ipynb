## FizzBuzz programming challenge

Write a short program that __prints each number from 1 to 100 on a new line__. For each __multiple of 3, print "Fizz"__ instead of the number. For each __multiple of 5, print "Buzz"__ instead of the number. For numbers which are __multiples of both 3 and 5, print "FizzBuzz"__ instead of the number.

```python
# Python

x = 1

for i in range(100):
    b = x + i    # fizzbuzz number to iterate from 1 to 100
    fizz = (b % 3 == 0)
    buzz = (b % 5 == 0)
    if (fizz and buzz):
        print('FizzBuzz')
    elif fizz:
        print('Fizz')
    elif buzz:
        print('Buzz')
    else:
        print(b)
    
```
